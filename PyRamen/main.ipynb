```python
salesdf = salesdf.rename({'Menu_Item':'item'}, axis='columns')
salesdf.head(3)
```

```python
salesgroup = salesdf.groupby(['item']).sum()
salesgroup
# menudf
```

```python
# adding price and cost columns from menudf to salesgroup

for row in menudf.values:
    item=row[0]
    price = row[-2]
    cost = row[-1]
    for line in salesgroup.index:
        if item == line:
            salesgroup['price']=price
            salesgroup['cost']=cost
            
salesgroup
    
```

```python
# adding revenue, cogs, and profit columns to salesgroup

salesgroup['revenue']=salesgroup['Quantity']*salesgroup['price']

salesgroup['cogs']=salesgroup['Quantity']*salesgroup['cost']

salesgroup['profit']=salesgroup['revenue']*salesgroup['cogs']

salesgroup
```
